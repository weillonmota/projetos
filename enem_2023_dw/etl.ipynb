```python
import pandas as pd
import time # Para cronometrar o tempo de carga

# Vamos usar a variável 'nome_do_arquivo' da Célula 1
# nome_do_arquivo = '../dados_brutos/MICRODADOS_ENEM_2023.csv' 

# --- 1. DEFINIÇÃO DAS COLUNAS (COM OS NOMES CORRIGIDOS) ---
colunas_para_carregar = [
    'NU_INSCRICAO',
    'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO',
    'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT', 'TP_STATUS_REDACAO',
    
    # --- COLUNAS DE LOCALIZAÇÃO CORRIGIDAS ---
    # Usando local da PROVA como substituto para local de RESIDÊNCIA
    'NO_MUNICIPIO_PROVA',  # <-- MUDANÇA (antes: NO_MUNICIPIO_RESIDENCIA)
    'SG_UF_PROVA',         # <-- MUDANÇA (antes: SG_UF_RESIDENCIA)
    
    # A coluna 'TP_LOCALIZACAO_RESIDENCIA' não existe e foi REMOVIDA
    
    'TP_SEXO', 'TP_FAIXA_ETARIA', 'TP_COR_RACA',
    'TP_ESCOLA', 'Q006'
]

print(f"Definidas {len(colunas_para_carregar)} colunas para carregar.")
print(f"Iniciando a carga do arquivo: '{nome_do_arquivo}'...")
print("(Isso pode levar alguns minutos...)")

# Medindo o tempo
start_time = time.time()

try:
    # --- 2. CARGA OTIMIZADA ---
    df_enem_completo = pd.read_csv(
        nome_do_arquivo, 
        encoding='latin-1', 
        sep=';',
        usecols=colunas_para_carregar # Usando a lista corrigida
    )
    
    end_time = time.time()
    
    print(f"\n--- SUCESSO! ---")
    print(f"Arquivo completo carregado em {end_time - start_time:.2f} segundos.")
    print("DataFrame 'df_enem_completo' está pronto.")
    
    # --- 3. MOSTRAR OS DADOS ---
    print("\n--- Amostra (5 primeiras linhas dos dados carregados): ---")
    display(df_enem_completo.head())
    
    print("\n--- Informações (Tipos de dados e valores nulos): ---")
    df_enem_completo.info()

except Exception as e:
    print(f"\n--- ERRO DURANTE A CARGA ---")
    print(f"Erro: {e}")
    print("Se o erro persistir, verifique os nomes da lista 'colunas_para_carregar' novamente.")
```

```python
# --- CÉLULA 4: Limpeza e Transformação (Filtrando Presentes) ---

print("Iniciando a limpeza dos dados...")

# Verificação inicial
if 'df_enem_completo' not in locals():
    print("ERRO: 'df_enem_completo' não foi encontrado.")
    print("Por favor, rode a Célula 3 (carga dos dados) primeiro.")
else:
    print(f"Total de inscritos (linhas) no DataFrame original: {len(df_enem_completo)}")

    # 1. Filtrando quem estava PRESENTE em todas as provas
    #    (TP_PRESENCA == 1 significa 'Presente')
    filtro_presenca = (
        (df_enem_completo['TP_PRESENCA_CN'] == 1) &
        (df_enem_completo['TP_PRESENCA_CH'] == 1) &
        (df_enem_completo['TP_PRESENCA_LC'] == 1) &
        (df_enem_completo['TP_PRESENCA_MT'] == 1)
    )

    # 2. Filtrando quem teve a REDAÇÃO VÁLIDA (Status 1 = "Sem Problemas")
    #    Isso elimina notas zero por fuga ao tema, cópia, etc.
    filtro_redacao = (df_enem_completo['TP_STATUS_REDACAO'] == 1)

    # 3. Criando o novo DataFrame LIMPO
    #    Usamos .copy() para evitar avisos de SettingWithCopyWarning no futuro
    df_enem_presentes = df_enem_completo[filtro_presenca & filtro_redacao].copy()

    # 4. Verificando o resultado (Respondendo à Pergunta 11)
    total_inscritos = len(df_enem_completo)
    total_presentes = len(df_enem_presentes)
    total_ausentes_ou_eliminados = total_inscritos - total_presentes
    
    print(f"\n--- Resposta à Pergunta 11 ---")
    print(f"Total de alunos PRESENTES e com redação válida: {total_presentes}")
    print(f"Total de alunos AUSENTES ou eliminados: {total_ausentes_ou_eliminados}")

    print("\n--- Amostra do DataFrame limpo (df_enem_presentes.head()): ---")
    # Note que agora não devem mais existir 'NaN' nas notas
    display(df_enem_presentes.head())

    print("\n--- Verificando valores nulos nas notas (DEPOIS da limpeza): ---")
    # Idealmente, o resultado aqui será '0' para todas as notas,
    # pois filtramos apenas quem estava presente.
    print(df_enem_presentes[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].isnull().sum())
```

```python
# --- CÉLULA 5: Transformação (Enriquecimento Geográfico) ---

print("Iniciando a Etapa de Transformação (T)...")
print("Criando colunas 'NO_REGIAO' (Pergunta 1) e 'TP_CAPITAL' (Pergunta 8).")

# 1. Dicionário de Mapeamento de Estados (UF) para Regiões
#    (Chave = UF, Valor = Região)
mapa_regioes = {
    'AC': 'Norte', 'AP': 'Norte', 'AM': 'Norte', 'PA': 'Norte', 'RO': 'Norte', 'RR': 'Norte', 'TO': 'Norte',
    'AL': 'Nordeste', 'BA': 'Nordeste', 'CE': 'Nordeste', 'MA': 'Nordeste', 'PB': 'Nordeste', 'PE': 'Nordeste', 'PI': 'Nordeste', 'RN': 'Nordeste', 'SE': 'Nordeste',
    'DF': 'Centro-Oeste', 'GO': 'Centro-Oeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste',
    'ES': 'Sudeste', 'MG': 'Sudeste', 'RJ': 'Sudeste', 'SP': 'Sudeste',
    'PR': 'Sul', 'RS': 'Sul', 'SC': 'Sul'
}

# 2. Lista de Capitais Brasileiras (para Pergunta 8)
#    Atenção aos acentos! Eles devem bater com os dados.
lista_capitais = [
    'Rio Branco', 'Maceió', 'Macapá', 'Manaus', 'Salvador', 'Fortaleza', 'Brasília', 'Vitória', 'Goiânia', 
    'São Luís', 'Cuiabá', 'Campo Grande', 'Belo Horizonte', 'Belém', 'João Pessoa', 'Curitiba', 'Recife', 
    'Teresina', 'Rio de Janeiro', 'Natal', 'Porto Alegre', 'Porto Velho', 'Boa Vista', 'Florianópolis', 
    'São Paulo', 'Aracaju', 'Palmas'
]

try:
    # --- Criando a coluna NO_REGIAO (Pergunta 1) ---
    # Usamos .map() para "traduzir" a UF para Região
    df_enem_presentes['NO_REGIAO'] = df_enem_presentes['SG_UF_PROVA'].map(mapa_regioes)
    
    # --- Criando a coluna TP_CAPITAL (Pergunta 8) ---
    # Usamos .apply() e uma função lambda para verificar se o município está na lista
    df_enem_presentes['TP_CAPITAL'] = df_enem_presentes['NO_MUNICIPIO_PROVA'].apply(
        lambda municipio: 'Capital' if municipio in lista_capitais else 'Interior'
    )

    print("\n--- SUCESSO! Colunas 'NO_REGIAO' e 'TP_CAPITAL' criadas. ---")
    
    # --- Verificação ---
    print("\nAmostra dos dados com as novas colunas:")
    # Mostra as colunas que usamos e as que criamos
    display(df_enem_presentes[['SG_UF_PROVA', 'NO_REGIAO', 'NO_MUNICIPIO_PROVA', 'TP_CAPITAL']].head())
    
    print("\nContagem de alunos por Região (verificando 'NO_REGIAO'):")
    print(df_enem_presentes['NO_REGIAO'].value_counts())
    
    print("\nContagem de alunos por Capital/Interior (verificando 'TP_CAPITAL'):")
    print(df_enem_presentes['TP_CAPITAL'].value_counts())

except Exception as e:
    print(f"\nERRO ao tentar criar as colunas de localização: {e}")
    print("Verifique se o DataFrame 'df_enem_presentes' existe.")
```

```python
# --- CÉLULA 6: Checkpoint (Salvando DataFrame Limpo) ---

print("Iniciando o salvamento do DataFrame 'df_enem_presentes'...")
print("Isso cria um checkpoint permanente do nosso trabalho de ETL.")

# O arquivo será salvo na mesma pasta onde o seu notebook está rodando.
# (No seu caso: ...\Projeto DW\enem_2023_presentes.parquet)
nome_arquivo_checkpoint = 'enem_2023_presentes.parquet'

try:
    # Tentando salvar em formato Parquet
    df_enem_presentes.to_parquet(nome_arquivo_checkpoint, index=False)
    
    print(f"\n--- SUCESSO! ---")
    print(f"DataFrame salvo com sucesso no arquivo: '{nome_arquivo_checkpoint}'")
    print("\nSe o seu notebook travar, você pode pular as células 1-5")
    print(f"e carregar este arquivo limpo com: pd.read_parquet('{nome_arquivo_checkpoint}')")

except ImportError:
    print("\n--- ERRO: Biblioteca 'pyarrow' não encontrada ---")
    print("O formato Parquet precisa de uma biblioteca auxiliar.")
    print("Por favor, rode o comando abaixo em uma nova célula e tente novamente:")
    print("!pip install pyarrow")
except Exception as e:
    print(f"\n--- ERRO ao tentar salvar o arquivo ---")
    print(f"Erro: {e}")
```

```python
# --- CÉLULA 7: Modelagem do Star Schema - dim_localizacao ---

print("Iniciando a modelagem do Star Schema...")
print("Criando a 'dim_localizacao' (Dimensão ONDE)...")

# 1. Selecionar as colunas relevantes
colunas_localizacao = [
    'NO_MUNICIPIO_PROVA', 
    'SG_UF_PROVA', 
    'NO_REGIAO', 
    'TP_CAPITAL'
]
dim_localizacao = df_enem_presentes[colunas_localizacao]

# 2. Remover duplicatas
print(f"Linhas antes de remover duplicatas: {len(dim_localizacao)}")
dim_localizacao = dim_localizacao.drop_duplicates().reset_index(drop=True)
print(f"Linhas após remover duplicatas (total de locais únicos): {len(dim_localizacao)}")

# 3. Criar a Chave Primária (ID_Local)
#    Vamos usar o índice (que agora vai de 0 até N-1)
#    É uma boa prática começar IDs de 1 (e não 0)
dim_localizacao['ID_Local'] = dim_localizacao.index + 1

# 4. Reorganizar as colunas (colocando a chave primária na frente)
colunas_finais = [
    'ID_Local', 
    'NO_MUNICIPIO_PROVA', 
    'SG_UF_PROVA', 
    'NO_REGIAO', 
    'TP_CAPITAL'
]
dim_localizacao = dim_localizacao[colunas_finais]

print("\n--- SUCESSO! 'dim_localizacao' criada. ---")
print("Amostra da Tabela de Dimensão 'dim_localizacao':")
display(dim_localizacao.head())

print("\nInformações da 'dim_localizacao':")
dim_localizacao.info()
```

### Modelagem do Data Warehouse (Star Schema)

Abaixo está o diagrama do nosso Star Schema. Estamos usando:
- **`dim_aluno`** (QUEM)
- **`dim_localizacao`** (ONDE)
- **`fato_notas`** (O QUÊ)

![Nosso Star Schema do ENEM](schema_enem.png)

```python
# --- CÉLULA 8: Modelagem do Star Schema - dim_aluno ---

print("Continuando a modelagem...")
print("Criando a 'dim_aluno' (Dimensão QUEM)...")

try:
    # 1. Selecionar as colunas do aluno
    colunas_aluno = [
        'NU_INSCRICAO',
        'TP_SEXO', 
        'TP_FAIXA_ETARIA', 
        'TP_COR_RACA', 
        'TP_ESCOLA', 
        'Q006'
    ]
    # Usamos .copy() para garantir que estamos criando um novo DataFrame
    dim_aluno = df_enem_presentes[colunas_aluno].copy()

    # 2. Resetar o índice (para garantir que ele seja de 0 a N-1)
    dim_aluno = dim_aluno.reset_index(drop=True)

    # 3. Criar a Chave Primária Surrogada (ID_Aluno)
    #    (Começando de 1)
    dim_aluno['ID_Aluno'] = dim_aluno.index + 1

    # 4. Renomear a Chave Natural (NU_INSCRICAO) para clareza
    #    'NK' = Natural Key (Chave do negócio)
    dim_aluno = dim_aluno.rename(columns={'NU_INSCRICAO': 'NK_Inscricao'})

    # 5. Reorganizar as colunas (colocando a chave primária na frente)
    colunas_finais_aluno = [
        'ID_Aluno',         # Chave Primária (Surrogada)
        'NK_Inscricao',     # Chave Natural (Business Key)
        'TP_SEXO', 
        'TP_FAIXA_ETARIA', 
        'TP_COR_RACA', 
        'TP_ESCOLA', 
        'Q006'
    ]
    dim_aluno = dim_aluno[colunas_finais_aluno]

    print(f"\n--- SUCESSO! 'dim_aluno' criada. ---")
    print(f"Total de alunos na dimensão: {len(dim_aluno)}")
    print("Amostra da Tabela de Dimensão 'dim_aluno':")
    display(dim_aluno.head())

    print("\nInformações da 'dim_aluno':")
    dim_aluno.info()

except Exception as e:
    print(f"\n--- ERRO ao criar a dim_aluno ---")
    print(f"Erro: {e}")
    print("Verifique se o DataFrame 'df_enem_presentes' existe.")
```

```python
# --- CÉLULA 9: Modelagem do Star Schema - fato_notas ---

print("Modelagem final: Criando a 'fato_notas' (Centro da Estrela)...")
print("(Este passo pode demorar alguns minutos, pois está juntando as tabelas)")

try:
    # 1. Começar com o DataFrame principal (que tem as notas e as chaves naturais)
    fato_notas = df_enem_presentes.copy()

    # --- Juntando as Chaves Estrangeiras (FKs) ---

    # 2. Trazer a FK 'ID_Aluno' da 'dim_aluno'
    #    Vamos juntar 'fato_notas' com 'dim_aluno'
    #    usando a chave natural 'NU_INSCRICAO' / 'NK_Inscricao'
    fato_notas = pd.merge(
        fato_notas, 
        dim_aluno[['NK_Inscricao', 'ID_Aluno']],
        left_on='NU_INSCRICAO',
        right_on='NK_Inscricao',
        how='left' # 'left' join para garantir que não perdemos nenhum aluno
    )

    # 3. Trazer a FK 'ID_Local' da 'dim_localizacao'
    #    Vamos juntar 'fato_notas' com 'dim_localizacao'
    #    usando as colunas de localização
    colunas_join_local = ['NO_MUNICIPIO_PROVA', 'SG_UF_PROVA', 'NO_REGIAO', 'TP_CAPITAL']
    fato_notas = pd.merge(
        fato_notas,
        dim_localizacao[['ID_Local'] + colunas_join_local],
        on=colunas_join_local,
        how='left' # 'left' join
    )

    print("Join das chaves estrangeiras concluído.")

    # --- Limpeza Final ---

    # 4. Selecionar APENAS as Fatos e as Chaves Estrangeiras (FKs)
    colunas_fato = [
        'ID_Aluno',     # FK para dim_aluno
        'ID_Local',     # FK para dim_localizacao
        'NU_NOTA_CN',   # Fato
        'NU_NOTA_CH',   # Fato
        'NU_NOTA_LC',   # Fato
        'NU_NOTA_MT',   # Fato
        'NU_NOTA_REDACAO' # Fato
    ]
    fato_notas = fato_notas[colunas_fato]

    # 5. Criar a Chave Primária (PK) da Tabela Fato
    fato_notas = fato_notas.reset_index(drop=True)
    fato_notas['ID_Fato'] = fato_notas.index + 1

    # 6. Reorganizar para o padrão (PK primeiro)
    fato_notas = fato_notas[['ID_Fato', 'ID_Aluno', 'ID_Local', 'NU_NOTA_CN', 
                             'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']]

    print(f"\n--- SUCESSO! 'fato_notas' criada. ---")
    print(f"Total de fatos (notas de alunos): {len(fato_notas)}")
    print("Amostra da Tabela Fato 'fato_notas':")
    display(fato_notas.head())

    print("\nInformações da 'fato_notas':")
    fato_notas.info()

except Exception as e:
    print(f"\n--- ERRO ao criar a fato_notas ---")
    print(f"Erro: {e}")
    print("Verifique se 'df_enem_presentes', 'dim_aluno' e 'dim_localizacao' existem.")
```
